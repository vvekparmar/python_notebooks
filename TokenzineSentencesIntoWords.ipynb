```python
import numpy
import nltk
```

```python
# default tokenizer used by nltk is 'Puntk' tokenizer,Punkt is unsupervised pretrained tokenizer. 
nltk.download('punkt')
```

```python
# importing nltk tokenize for words and sentences.
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
text = "These methods are generally used while doing the pre-processing step. These methods select features from the dataset irrespective of the use of any machine learning algorithm. In terms of computation, they are very fast and inexpensive and are very good for removing duplicated, correlated, redundant features but these methods do not remove multicollinearity. Selection of feature is evaluated individually which can sometimes help when features are in isolation (don’t have a dependency on other features) but will lag when a combination of features can lead to increase in the overall performance of the model."
```

```python
# tokenize a sentence
sent_tokens = sent_tokenize(text)
print(sent_tokens)
```

```python
# tokenize a words
word_tokens = word_tokenize(text)
print(word_tokens)
```

```python
print(len(word_tokens))
```

```python
# import punkt sentence tokenizer
from nltk.tokenize.punkt import PunktSentenceTokenizer
```

```python
# init object
pst = PunktSentenceTokenizer()
```

```python
# tokenize the sentence
sent_tokenize = pst.tokenize(text)
print(sent_tokenize)
```

```python
# pst has span_tokenize() for printing out character ranges in result tokenize
span_tokens = pst.span_tokenize(text)
print(span_tokens)
```

```python
sentences = pst.sentences_from_tokens(word_tokens)
print(list(sentences))
```

```python
# WhitespaceTokenizer from nltk
from nltk.tokenize import WhitespaceTokenizer
```

```python
# init WhitespaceTokenizer object 
wt = WhitespaceTokenizer()
```

```python
# tokenize words
word_tokens = wt.tokenize(text)
print(word_tokens)
```

```python
# check frequency distribution of words using nltk.probability
from nltk.probability import FreqDist
```

```python
freq_dist = FreqDist(word_tokens)
print(freq_dist)
```

```python
# finding most common words from freq_dist
freq_dist.most_common(20)
```

```python
# Frequency Distribution of particular word
freq_dist.freq('the')
```

```python
freq_dist.freq('select')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10,8))
freq_dist.plot(20, cumulative=False)
plt.show()
```

```python
# converts text into lower case for cleaning purpose
text_contents = text.lower()
word_tokens = wt.tokenize(text_contents)
print(word_tokens)
```

```python
# download for 'stopwords' corpus for removing them
nltk.download('stopwords')
```

```python
# import stopwords corpus
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', 'The'])
print(stop_words)
```

```python
# filter out the word, that are not stopwords
filtered_words = []
for w in word_tokens:
  if w not in stop_words:
    filtered_words.append(w)
```

```python
print(filtered_words)
```

```python
# FREQUENCY DISTRIBUTION of 'filtered_words'
freq_dist = FreqDist(filtered_words)
print(freq_dist)
```

```python
freq_dist.most_common(20)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
freq_dist.plot(20, cumulative=False)
plt.show()
```
