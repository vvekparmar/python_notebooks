```python
import sklearn
```

```python
# from sklearn import CountVectorizer for conversion
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = ["These methods are generally used while doing the pre-processing step.",
        "These methods select features from the dataset irrespective of the use of any machine learning algorithm.",
        "In terms of computation, they are very fast and inexpensive and are very good.",
        "for removing duplicated, correlated, redundant features but these methods do not remove multicollinearity.",
        "Selection of feature is evaluated individually which can sometimes help when features are in isolation.",
        "donâ€™t have a dependency on other features but will lag.",
        "when a combination of features can lead to increase in the overall performance of the model."]
```

```python
# init instance of CountVectorizer
count_vectorizer = CountVectorizer()
```

```python
# fit our text in count_vectorizer
count_vectorizer.fit(text)
```

```python
# get tokens, extracted from CountVectorizer
count_vectorizer.get_feature_names_out()
```

```python
# get stop words from the count_vectorizer
count_vectorizer.get_stop_words()
```

```python
# count_vectorizer has vocabulary
count_vectorizer.vocabulary_
```

```python
# get specific index from the word in count_vectorizer
count_vectorizer.vocabulary_.get("select")
```

```python
# transform vector into bag of words representation
transformed_vector = count_vectorizer.transform(text)
transformed_vector.shape
```

```python
print(transformed_vector.toarray())
```

```python
# Now try to check some sentence's words are not presented in vocabulary
test_case = ["Hi, How Your days going till now!"]
```

```python
transformed_vector = count_vectorizer.transform(test_case)
```

```python
print(transformed_vector.toarray())
```

```python
# we have to fit count_vectorizer in this way
count_vectorizer.fit(text + test_case)
```
