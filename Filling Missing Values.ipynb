```python
import pandas as pd
import numpy as np
```

```python
# here np.nan meaning is missing values, created for demonstration purpose
toy_df = pd.DataFrame({"numbers" : [1, 2, 3, np.nan, np.nan, np.nan, 7]})
toy_df
```

```python
# using isnull() is check for null or missing values, sum() summation of missing values column 
toy_df.isnull().sum()
```

```python
# dropna() is used to drop all null or missing values
dropped_toy_df = toy_df.dropna()
dropped_toy_df
```

```python
# fillna() is used to fill missing values with particular default value
filled_toy_df = toy_df.fillna(0)
filled_toy_df
```

```python
# backfill technique to fill missing values, Backfill uses after value from missing value
backfilled_toy_df = toy_df.fillna(method='bfill')
backfilled_toy_df
```

```python
# forwardfill technique to fill missing values, Forwardfill uses after value from missing value
forwardfill_toy_df = toy_df.fillna(method='ffill')
forwardfill_toy_df
```

```python
# meanfilled or average value using mean()
meanfilled_toy_df = toy_df.fillna(toy_df['numbers'].mean())
meanfilled_toy_df
```

```python
# pd provides interpolation technique, estimates values that should use in missing fields
# interpolation is a mathematical technique where you construct new data points within the range of a discrete set of known data points
interpolated_toy_df = toy_df.interpolate()
interpolated_toy_df
```
