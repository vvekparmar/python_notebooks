**Method Overloading**

```python
class Addition :
    def addNumbers(self, a = None, b = None, c = None):
        if (a is not None) and (b is not None) and (c is not None):
          print("Addition : {}".format(a+b+c))

        elif (a is not None) and (b is not None):
          print("Addition : {}".format(a+b))

        else:
          print("Addition : {}".format(a+a))

obj = Addition()
obj.addNumbers(10)
obj.addNumbers(10,20)
obj.addNumbers(10,20,30)
```

**Inheritance & Method Overridding**

```python
class Parent:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def printData(self):
    print("Name : {}".format(self.name))
    print("Age : {}".format(self.age))
```

**Recursion**

```python
#Factorial

def fact(n):
  if(n == 0 or n == 1):
    return n
  return n * fact(n-1)

print("Factorial : {}".format(fact(7)))
```

```python
#Fibonacci Series

def fibo(n):
  if n <= 1:
    return n
  else:
    return(fibo(n-1) + fibo(n-2))

n = 10
for i in range(n):
	print(fibo(i))
```

**Abstract Class**


```python
from abc import ABC, abstractmethod

class Polygon(ABC):
	@abstractmethod
	def noOfSides(self):
		pass

class Triangle(Polygon):
  def noOfSides(self):
    print("I have 3 Sides.")

class Quadrilateral(Polygon):
  def noOfSides(self):
    print("I have 4 Sides")

class Pentagon(Polygon):
  def noOfSides(self):
    print("I have 5 Sides")

t = Triangle()
t.noOfSides()

q = Quadrilateral()
q.noOfSides()

p = Pentagon()
p.noOfSides()
```

```python
# Python program showing
# abstract properties

import abc
from abc import ABC, abstractmethod

class parent(ABC):
	@abc.abstractproperty
	def geeks(self):
		return "parent class"

class child(parent):
	@property
	def geeks(self):
		return "child class"

try:
	r = child()
	print(r.geeks)
except Exception as err:
	print (err)

r = child()
print (r.geeks)
```

**Decorator** 

```python
#Divide Function Decorator
def updateDivide(func):
    def inner(a, b):
        if b == 0:
            print("Denominator must be > 0")
            return
        return func(a, b)
    return inner

@updateDivide
def divide(a, b):
    print(a/b)
```

```python
divide(2,5)
```

**Yield Keyword**

```python
# generator to print even numbers
def printEven(test_list) :
	for i in test_list:
		if i % 2 == 0:
			yield i

# initializing list
test_list = [1, 4, 5, 6, 7]

# printing initial list
print ("The original list is : {}".format(test_list))

# printing even numbers
print ("The even numbers in list are : ", end = " ")
for j in printEven(test_list):
	print (j, end = " ")
```

```python
# A Python program to generate squares from 1
# to 100 using yield and therefore generator

def nextSquare():
	i = 1
	while True:
		yield i*i				
		i += 1

for num in nextSquare():
	if num > 100:
		break
	print(num)
```

**Combinatoric Iterators**

```python
#Product Iterator
from itertools import product

print("The cartesian product using repeat:")
print(list(product([1, 2], repeat = 2)))
print()

print("The cartesian product of the containers:")
print(list(product(['ABC', 'DEF', 'XYZ'], '2')))
print()

print("The cartesian product of the containers:")
print(list(product('AB', [3, 4])))
```

```python
#Permutation Iterator
from itertools import permutations

print ("All the permutations of the given list is:")
print (list(permutations([1, 'abc'], 2)))
print()

print ("All the permutations of the given string is:")
print (list(permutations('AB')))
print()

print ("All the permutations of the given container is:")
print(list(permutations(range(3), 2)))
```

```python
#Combination Iterator
from itertools import combinations

print ("All the combination of list in sorted order(without replacement) is:")
print(list(combinations(['A', 2], 2)))
print()

print ("All the combination of string in sorted order(without replacement) is:")
print(list(combinations('AB', 2)))
print()

print ("All the combination of list in sorted order(without replacement) is:")
print(list(combinations(range(2), 1)))
```

```python
#combinations_with_replacement Iterator
from itertools import combinations_with_replacement

print ("All the combination of string in sorted order(with replacement) is:")
print(list(combinations_with_replacement("AB", 2)))
print()

print ("All the combination of list in sorted order(with replacement) is:")
print(list(combinations_with_replacement([1, 2], 2)))
print()

print ("All the combination of container in sorted order(with replacement) is:")
print(list(combinations_with_replacement(range(2), 1)))
```
