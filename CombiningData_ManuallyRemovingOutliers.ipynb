```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mall_customers_info = pd.read_csv("/content/Mall_Customers.csv")
mall_customers_info.head()
```

```python
mall_customers_info.columns
```

```python
# check shape for the dataframe
mall_customers_info.shape
```

```python
# create new dataframe for scores
mall_customers_scores = mall_customers_info[['CustomerID','Spending Score (1-100)']]
mall_customers_scores.head()
```

```python
mall_customers_info = mall_customers_info.drop(['Spending Score (1-100)'], axis=1)
mall_customers_info.head()
```

```python
print("Shape of mall_customers_info :", mall_customers_info.shape)
print("Shape of mall_customers_scores :", mall_customers_scores.shape)
```

```python
# Merging both the dataframes
customer_data = pd.merge(mall_customers_info, mall_customers_scores[['CustomerID', 'Spending Score (1-100)']], on="CustomerID")
customer_data.head()
```
