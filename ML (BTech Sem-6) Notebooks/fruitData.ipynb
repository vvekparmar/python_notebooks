```python
#from scikit-learn library import tree classifier
from sklearn import tree
```

```python
#features of data
features = [[150,0],[190,0],[140,1],[120,1]] #0-bumpy and 1-smooth

#labels of data
labels = [0,0,1,1] #0-orange and 1-apple
```

```python
#create object of tree classifier
classifier = tree.DecisionTreeClassifier()
```

```python
#create a model using fit function of tree classifier
model = classifier.fit(features,labels)
```

```python
#predict data using our model
print(model.predict([[190,1]]))

#if it print '1' then it classify as Apple, and if '0' then it classify as Orange 
```
