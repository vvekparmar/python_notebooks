```python
survivedPeople = data.Survived.value_counts()
print(survivedPeople)
```

```python
plotSurvivedData = survivedPeople.plot(kind='bar',color=['red','green'])
plotSurvivedData.set_xlabel("Survived or Not")
plotSurvivedData.set_ylabel("Passenger Counts")
```

```python
#class based survivibility
plotPClass = data.Pclass.value_counts().plot(kind='bar',title='class based count',color=['red','green','blue'])
plotPClass.set_xlabel("PClass")
plotPClass.set_ylabel("Total Passenger")
```

```python
countGroupSurvivedPClass = data[['Pclass','Survived']].groupby('Pclass').count()
pltGroupSurvivedPClass = data[['Pclass','Survived']].groupby('Pclass').mean().Survived.plot(kind='bar')
pltGroupSurvivedPClass.set_xlabel('PCLass')
pltGroupSurvivedPClass.set_ylabel('Survival Probability')
#higher mean indicated class 1 had higher chance of survive
```

```python
pltGenderSurvivedPClass = data[['Sex','Survived']].groupby('Sex').mean().Survived.plot(kind='bar')
pltGenderSurvivedPClass.set_xlabel('sex')
pltGenderSurvivedPClass.set_ylabel('Survival Probability')
```
