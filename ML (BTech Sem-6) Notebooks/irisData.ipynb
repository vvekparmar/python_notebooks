```python
#load iris dataset from scikit-learn library
from sklearn.datasets import load_iris
```

```python
#create object of dataset
data = load_iris()
```

```python
#import train_test_split from scikit-learn library for splitting a data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.10,random_state=42)
```

```python
#import tree classifier from scikit-learn library
from sklearn import tree
```

```python
#create object of tree classifier
classifier = tree.DecisionTreeClassifier()
```

```python
#create model using fit function of tree classifier
model = classifier.fit(X_train,y_train)
```

```python
#predict data using model
print("Predicted Data : ",model.predict(X_test))

#print target value or actual data from dataset
print("Actual Data    : ",y_test)
```
