```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import shutil
```

```python
def image_feature(direc):
    model = InceptionV3(weights='imagenet', include_top=False)
    features = [];
    img_name = [];
    for i in tqdm(direc):
        fname='clusterImage'+'/'+i
        img=image.load_img(fname,target_size=(224,224))
        x = img_to_array(img)
        x=np.expand_dims(x,axis=0)
        x=preprocess_input(x)
        feat=model.predict(x)
        feat=feat.flatten()
        features.append(feat)
        img_name.append(i)
    return features,img_name
```

```python
img_path = os.listdir('clusterImage')
img_features,img_name=image_feature(img_path)
```

```python
#Creating Clusters
k = 3
clusters = KMeans(k, random_state = 40)
clusters.fit(img_features)
```

```python
image_cluster = pd.DataFrame(img_name,columns=['image'])
image_cluster["labels"] = clusters.labels_

#if we view full dataframe then we use set_option function
#pd.set_option("display.max_rows", None, "display.max_columns", None) 

image_cluster # {0 -> 'horse', 1 -> 'cat', 2 -> 'dog'}
```
