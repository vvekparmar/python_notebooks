```
import numpy, random, os
```

```
lr = 1 #learning rate
bias = 1
```

```
#generating weights x0,x1,bias using random functions
weights = [random.random(),random.random(),random.random()]
```

```
print("Initial Weight x0 :- ",weights[0])
print("Initial Weight x1 :- ",weights[1])
print("Initial Weight bias :- ",weights[2])
```

```
def perceptron(input1, input2, desiredOutput,correctedPrediction):
    if(correctedPrediction == 4):
        return
    predictedOutput = input1 * weights[0] + input2 * weights[1] + bias * weights[2]
    if predictedOutput > 0.8: #activation function
        predictedOutput = 1
    else:
        predictedOutput = 0
    
    if(predictedOutput == desiredOutput):
        correctedPrediction += 1
        return
        
    error = desiredOutput - predictedOutput
    weights[0] += error * input1 * lr
    weights[1] += error * input2 * lr
    weights[2] += error * bias * lr
```

```
correctedPrediction = 0
for i in range(10) : #epoch value
    perceptron(1, 1, 1,correctedPrediction)
    perceptron(1, 0, 1,correctedPrediction)
    perceptron(0, 1, 1,correctedPrediction)
    perceptron(0, 0, 0,correctedPrediction) 
```

```
#take input and make predictions

x = int(input('Input 1 : ')) 
y = int(input('Input 2 : '))

predictedOutput = x * weights[0] + y * weights[1] + bias * weights[2]

print("Final Weight x0 :- ",weights[0])
print("Final Weight x1 :- ",weights[1])
print("Final Weight bias :- ",weights[2])

if predictedOutput > 0.8: #activation function
    predictedOutput = 1
else:
    predictedOutput = 0
    
print (x, " OR ", y, " is ",predictedOutput)
```
