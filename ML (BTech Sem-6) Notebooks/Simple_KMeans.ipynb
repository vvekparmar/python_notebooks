```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([[1, 2], [1, 4], [1, 0],[10, 2], [10, 4], [10, 0]])
kmeans = KMeans(n_clusters=2, random_state=10)
kmeans.fit(X)

print("Labels : ", kmeans.labels_)
print("Prediction : ", kmeans.predict([[100, 0], [1, 100]]))
print("Cluster Center : ", kmeans.cluster_centers_)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target

kmeans = KMeans(n_clusters=2, random_state=30).fit(X)

print("Labels : ", kmeans.labels_)
print("Labels Size : ", kmeans.labels_.size)
```

```python
print(X[5])
print(y[5])
```

```python
print(kmeans.predict([X[5]]))

print("Inertia : ", kmeans.inertia_)
```

```python
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(n_clusters=3, random_state=0,batch_size=5).fit(X)
print("Labels : ", kmeans.labels_)
print("Labels Size : ", kmeans.labels_.size)
```
