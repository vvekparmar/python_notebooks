```python
#import iris dataset from scikit-learn library
from sklearn.datasets import load_iris 
```

```python
#creating object of iris datasets
irisData = load_iris() 
```

```python
# Create feature and target arrays 
X = irisData.data 
y = irisData.target
```

```python
#import train_test_split from scikit-learn library
from sklearn.model_selection import train_test_split 

# Split into training and test set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) 
```

```python
#import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

#create scaler object using StandardScaler()
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#import KNeighbors from scikit-learn
#by default distance metric is minkowski
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=7,metric='minkowski') 
model = knn.fit(X_train, y_train) 
```

```python
#import KNeighbors from scikit-learn
#Euclidean Distance Metric
knn = KNeighborsClassifier(n_neighbors=7,metric='euclidean') 
model = knn.fit(X_train, y_train) 
```

```python
#import KNeighbors from scikit-learn
#Manhattan Distance Metric
knn = KNeighborsClassifier(n_neighbors=7,metric='manhattan') 
model = knn.fit(X_train, y_train) 
```

```python
#import KNeighbors from scikit-learn
#Chebyshev Distance Metric
knn = KNeighborsClassifier(n_neighbors=7,metric='chebyshev') 
model = knn.fit(X_train, y_train) 
```

```python
#import KNeighbors from scikit-learn
#Mahalanobis Distance Metric
import numpy as np

#np.cov() means covariance 
#If cov(xi, xj) = 0 then variables are uncorrelated
#If cov(xi, xj) > 0 then variables positively correlated
#If cov(xi, xj) > < 0 then variables negatively correlated

knn = KNeighborsClassifier(n_neighbors=7,metric='mahalanobis',metric_params={'V': np.cov(X_train[:4])}) 
model = knn.fit(X_train, y_train) 
```

```python
from sklearn.metrics import accuracy_score
predictedData = model.predict(X_test)
accuracy_score(y_test, predictedData) * 100
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictedData)
```
