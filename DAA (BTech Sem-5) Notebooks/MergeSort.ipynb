```
import numpy as np
```

```
array = np.random.randint(1,1000,size=30)
```

```
def merge(arr, left, middle, right): 
    n1 = middle - left + 1
    n2 = right - middle 
    leftHalves = [0] * (n1) 
    rightHalves = [0] * (n2) 
  
    for i in range(0 , n1): 
        leftHalves[i] = arr[left + i] 
  
    for j in range(0 , n2): 
        rightHalves[j] = arr[middle + 1 + j] 
  
    i = 0     
    j = 0     
    k = left     
  
    while i < n1 and j < n2 : 
        if leftHalves[i] <= rightHalves[j]: 
            arr[k] = leftHalves[i] 
            i += 1
        else: 
            arr[k] = rightHalves[j] 
            j += 1
        k += 1
  
    while i < n1: 
        arr[k] = leftHalves[i] 
        i += 1
        k += 1
  
    while j < n2: 
        arr[k] = rightHalves[j] 
        j += 1
        k += 1
```

```
 def mergeSort(arr,left,right): 
    if left < right: 
        middle = (left+(right-1))//2 
        mergeSort(arr, left, middle) 
        mergeSort(arr, middle+1, right) 
        merge(arr, left, middle, right) 
  
```

```
from datetime import datetime
```

```
#best case of merge sort of 30 elements
arr = np.arange(start=1,stop=30,step=1)
beforeTime = datetime.now()
mergeSort(arr,0,len(arr)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Best Case :- ",exeTime)
```

```
#average case of merge sort of 30 elements
beforeTime = datetime.now()
mergeSort(array,0,len(array)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Average Case :- ",exeTime)
```

```
#worse case of merge sort of 30 elements
arr = np.arange(start=30,stop=1,step=-1)
beforeTime = datetime.now()
mergeSort(arr,0,len(arr)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Worse Case :- ",exeTime)
```
