```
import numpy as np
```

```
array = np.random.randint(1,10,size=10)
```

```
def countingSort(array):
    size = len(array)
    output = [0] * size
    count = [0] * 10

    for i in range(0, size):
        count[array[i]] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]
    
    i = size - 1
    while i >= 0:
        output[count[array[i]] - 1] = array[i]
        count[array[i]] -= 1
        i -= 1

    for i in range(0, size):
        array[i] = output[i]
```

```
from datetime import datetime
```

```
#best case for counting sort
arr = np.arange(start=1,stop=10,step=1)
beforeTime = datetime.now();
countingSort(arr)
afterTime = datetime.now();
exeTime = afterTime-beforeTime
print(exeTime)
```

```
#average case for counting sort
beforeTime = datetime.now();
countingSort(array)
afterTime = datetime.now();
exeTime = afterTime-beforeTime
print(exeTime)
```

```
#worse case for counting sort
arr = np.arange(start=9,stop=0,step=-1)
beforeTime = datetime.now();
countingSort(arr)
afterTime = datetime.now();
exeTime = afterTime-beforeTime
print(exeTime)
```
