```
import numpy as np
```

```
array = np.random.randint(1,1000,size=30)
```

```
def partition(arr,low,high): 
    i = ( low-1 )         
    pivot = arr[high]     
  
    for j in range(low , high): 
        if   arr[j] < pivot:  
            i = i+1 
            arr[i],arr[j] = arr[j],arr[i] 
  
    arr[i+1],arr[high] = arr[high],arr[i+1] 
    return ( i+1 ) 
```

```
def quickSort(arr,low,high): 
    if low < high: 
        pi = partition(arr,low,high) 
        quickSort(arr, low, pi-1) 
        quickSort(arr, pi+1, high)
    return arr
```

```
from datetime import datetime
```

```
#best test case for quick sort 30 elements
arr = np.arange(start=1,stop=30,step=1)
beforeTime = datetime.now()
arr = quickSort(arr,0,len(arr)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Best Case :- " ,exeTime)
```

```
#average test case for quick sort 30 elements
beforeTime = datetime.now()
arr = quickSort(array,0,len(arr)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Average Case :- " ,exeTime)
```

```
#worse test case for quick sort 30 elements
arr = np.arange(start=30,stop=1,step=-1)
beforeTime = datetime.now()
arr = quickSort(arr,0,len(arr)-1)
afterTime = datetime.now()
exeTime = afterTime-beforeTime
print("Worse Case :- " ,exeTime)
```
