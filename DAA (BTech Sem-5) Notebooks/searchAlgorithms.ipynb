```
def linearIterative(array, ele):
   for i in range(len(array)):
      if array[i] == ele:
         return i+1
   return -1
```

```
def linearRecursive(array, low, high, ele): 
    if high < low: 
        return -1
    if array[low] == ele: 
        return low 
    if array[high] == ele: 
        return high 
    return linearRecursive(array, low+1, high-1, ele)  
```

```
def binaryRecursion(arr, low, high, x): 
    if high >= low: 
        mid = (high + low) // 2
        if arr[mid] == x: 
            return mid 
        elif arr[mid] > x: 
            return binaryRecursion(arr, low, mid - 1, x) 
        else: 
            return binaryRecursion(arr, mid + 1, high, x) 
    else: 
        return -1
  
```

```
def binaryIterative(arr, x): 
    low = 0
    high = len(arr) - 1
    mid = 0
    while low <= high: 
        mid = (high + low) // 2
        if arr[mid] < x: 
            low = mid + 1 
        elif arr[mid] > x: 
            high = mid - 1
        else: 
            return mid
    return -1
  
```
