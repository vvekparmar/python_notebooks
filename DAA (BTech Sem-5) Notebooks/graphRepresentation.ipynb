```
import numpy as np
```

```
#graph represenation using Adjacency Matrix.
class AdjacencyMatrix:
  def __init__(self,vertices):
    self.vertices = vertices
    self.matrix = np.zeros([vertices,vertices],dtype=int)
  def addEdges(self,u,v,weight):
    self.matrix[u][v] = weight
    self.matrix[v][u] = weight
```

```
graphMatrix = AdjacencyMatrix(8)
```

```
graphMatrix.addEdges(0,3,7)
graphMatrix.addEdges(0,4,1)
graphMatrix.addEdges(0,1,7)
graphMatrix.addEdges(3,5,3)
graphMatrix.addEdges(1,5,5)
graphMatrix.addEdges(5,6,6)
graphMatrix.addEdges(5,2,2)
graphMatrix.addEdges(1,6,8)
```

```
#Graph representation using Adjacency List.
class AdjacencyList:
  def __init__(self):
    self.vertices = 0
    self.verticesNo = 0
    self.resultList = {}

  def addVertex(self,v):
    if v in self.resultList:
      print("Vertex ", v , " already exists.")
    else:
      self.verticesNo = self.verticesNo + 1
      self.resultList[v] = []

  def addEdge(self,src, dest, weight):
    if src not in self.resultList:
      print("Vertex ", src, " does not exist.")
    elif dest not in self.resultList:
      print("Vertex ", dest, " does not exist.")
    else:
      self.resultList[src].append([dest,weight]) #directed weighted graph
      #self.resultList[dest].append([src,weight]) #undirected weighted graph
      #self.resultList[src].append(dest) #directed unweighted graph
      #self.resultList[dest].append(src) #undirected unweighted graph
  
  def traverseGraph(self):
    for vertex in self.resultList:
      for edges in self.resultList[vertex]:
        print(vertex, " -> ", edges[0], " weight: ", edges[1])
```

```
#add vertex in list
graphList = AdjacencyList() 
graphList.addVertex(0)
graphList.addVertex(1)
graphList.addVertex(2)
graphList.addVertex(3)
graphList.addVertex(4)
graphList.addVertex(5)
graphList.addVertex(6)
graphList.addVertex(7)
```

```
#add Edges
graphList.addEdge(0,3,5)
graphList.addEdge(0,4,8)
graphList.addEdge(0,2,1)
graphList.addEdge(3,5,1)
graphList.addEdge(3,7,9)
graphList.addEdge(5,7,4)
graphList.addEdge(2,6,6)
graphList.addEdge(4,6,3)
```
