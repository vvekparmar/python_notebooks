```
import numpy as np
```

```
arrayT = np.random.randint(1,1000,size=1000)
```

```
arrayTT = np.random.randint(1,1000,size=10000)
```

```
arrayL = np.random.randint(1,1000,size=100000)
```

```
def bubbleSort(array):
    length = len(array)
    for i in np.arange(length-1):
        for j in np.arange(length-i-1):
            if(array[j] > array[j+1]):
                array[j],array[j+1] = array[j+1],array[j]
    return array
```

```
def selectionSort(array):
    length = len(array)
    for i in np.arange(length):
        min = i
        for j in np.arange(i+1,length):
            if(array[min] > array[j]):
                min = j
        array[i],array[min] = array[min],array[i]
    return array
```

```
def insertionSort(array):
  length = len(array)
  for i in np.arange(1,length):
    curr = array[i]
    pos = i
    while pos > 0 and array[pos-1] > curr:
      array[pos] = array[pos-1]
      pos = pos -1 
      array[pos] = curr
  return array
```
